```python
import requests
```

```python
URL = "https://aazk.org/job_postings/animal-attendant-3/"
page = requests.get(URL)

print(page.text)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content, "html.parser")
```

Position Title

```python
PositionTitle = soup.find(itemprop = 'headline')
print(PositionTitle.text)
```

```python

Facility = soup.find(class_="entry-content clearfix")
Facility1 = Facility
#Facility2 = Facility.find("Closing Date")
for x in Facility1:
    print(type(x))
    print(x.name)
    print(x.text)
#print(Facility.text)
Facility1
```
